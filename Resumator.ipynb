```python
import pandas as pd
import numpy as np
from weasyprint import HTML, CSS
from weasyprint.text.fonts import FontConfiguration
```

```python
mode = 1

if mode ==1:
    varopp = 'internship'
    varMS = 'Jan 2023'
elif mode == 2:
    varopp = 'full time'
    varMS = 'Sep 2022'

all_profiles = [
    'Unival',
#     'ML', # 0-2
#     'Analyst','DataEngineer','SDE', # 3-5
#     'Web',
#     'Industrial',
#     'DataScientist'
] # 6,7

df = pd.read_excel('Resume tabulated (copy).xlsx')

df.replace({'{varMS}': varMS, '{varopp}': varopp}, regex = True, inplace = True)

for profile in all_profiles:
# profile =  'Analyst'


    df_res_table = df.copy()
    
    
    df_res_table['list'].fillna(False, inplace=True)

    for md in all_profiles:
        df_res_table.loc[df_res_table[profile] == md, profile] = df_res_table[md][df_res_table[profile] == md] 
    
    df_res_table[profile].fillna(df_res_table['Unival'], inplace = True)
    
    
    df_res_table = df_res_table[df_res_table[profile] != 'Remove']
    

    content_cols = ['Section', 'Tenure','Heading', 'Subhead',profile]
    css_4 = """@page {
            size: A4; /* Change from the default size of A4 */
            margin: 0; /* Set margin on each page */
          }"""

    with open('res_simple_upper.html','r') as file:
        res_upper = file.read()

    html_to_add = """"""
    section_counter = 0
    ul_prev = False

    for idx,row in df_res_table.iterrows():

        if ul_prev and not row['list']:
            html_to_add += """</ul></div>""" # closing ul and resumecell


        if row.notnull().Section:
            if section_counter > 0:
                html_to_add += """\n        </div>""" # closing previous section
            html_to_add += """\n        <div class="section" id="section{}">
                <h3 class="h3 sectionhead">{}</h3>""".format(section_counter, row.astype(str).Section)
            section_counter +=1
            ul_prev = row['list']

            continue
        
        rscell = "resumecell"
        if row.notnull().Tenure:
            rscell += " grid-container" # making grid for tenure
            
        if not row['list']:
            html_to_add += """<div class="{}">""".format(rscell) # adding res cell helps tenure to attach cos relative position of resumecell
        
        if row['list'] and not ul_prev:
            html_to_add += """<div class="resumecell"><ul>""" # adding ul
            
        li_class = ""
        if row.notnull().Tenure:
            li_class += " grid-container" # making grid for tenure
        
        if row['list']:
            html_to_add += """<li class="{}">""".format(li_class) # adding li
        
        if row.notnull().Tenure:
            html_to_add += """<div class="grid-left-elem">""".format(row.Heading)
        
        if row.notnull().Heading:    # adding heading
            html_to_add += """<span><strong>{}</strong></span>""".format(row.Heading)
        
        if row.notnull()[profile]:
            html_to_add += """<span>{}</span>""".format(row[profile])
        
        if row.notnull().Tenure:
            html_to_add += """</div><div class="tenure grid-right-elem">{}</div>""".format(row.Tenure)
        
        if not row['list']: # normal case
            html_to_add += """</div>""" # class="resumecell"
        
        if row['list']:
            html_to_add += """</li>""" # closing li
        
        ul_prev = row['list']


    res_bottom = """</div></div></div>
    <div class="right"></div>
    <div id="footer"></div>
    </body></html>"""
    basic_resume = res_upper+ html_to_add +res_bottom

    font_config = FontConfiguration()
    html = HTML(string=basic_resume)
    css = CSS(string=css_4, font_config=font_config)
    html.write_pdf(
        'v2/resume_{}_{}.pdf'.format(profile, mode),
        stylesheets=[css],
        font_config=font_config
    )
    
#     break
```
